```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for folder in os.listdir("data"):
    path_to_folder = os.path.join("data", folder)
    y_original = pd.read_csv(os.path.join(path_to_folder, "y_original.csv"), header=None)
    y = pd.read_csv(os.path.join(path_to_folder, "y.csv"), header=None)
    gamma_c = pd.read_csv(os.path.join(path_to_folder, "gamma_c.csv"), header=None)
    gamma_python = pd.read_csv(os.path.join(path_to_folder, "gamma_python.csv"), header=None)
    
    fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
    
    ax[0][0].matshow(y_original)
    ax[0][0].set_title("y original")
    ax[0][1].matshow(y)
    ax[0][1].set_title("y noised")
    ax[1][0].matshow(gamma_c)
    ax[1][0].set_title("y denoised with c")
    ax[1][1].matshow(gamma_python)
    ax[1][1].set_title("y denoised with python")
    
    for i in range(2):
        for j in range(2):
            ax[i][j].set_xticks([])
            ax[i][j].set_yticks([])

    plt.suptitle(folder)
            
    plt.show()
```
