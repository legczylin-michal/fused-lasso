```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_original = pd.read_csv("c/data/y_original.csv", header=None)
y = pd.read_csv("c/data/y.csv", header=None)
gamma = pd.read_csv("c/data/gamma.csv", header=None)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
ax[0].matshow(y_original)
ax[1].matshow(y)
ax[2].matshow(gamma)
for i in range(3):
    ax[i].set_xticks([])
    ax[i].set_yticks([])
plt.show()
```
